```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/grapes_new.csv')
df
```

```python
label_encoder1 = LabelEncoder()
onehot_encoder = OneHotEncoder(sparse=False)
```

```python
onehot_encoder_shape = OneHotEncoder(sparse=False)
shape_encoded = onehot_encoder_shape.fit_transform(df[['SHAPE']])
shape_encoded_df = pd.DataFrame(shape_encoded, columns=onehot_encoder_shape.get_feature_names_out(['SHAPE']))

# Initialize OneHotEncoder for 'SOIL_TYPE'
onehot_encoder_soil = OneHotEncoder(sparse=False)
soil_type_encoded = onehot_encoder_soil.fit_transform(df[['SOIL_TYPE']])
soil_type_encoded_df = pd.DataFrame(soil_type_encoded, columns=onehot_encoder_soil.get_feature_names_out(['SOIL_TYPE']))

# Concatenate the original dataframe with the one-hot encoded columns
df_encoded = pd.concat([df, shape_encoded_df, soil_type_encoded_df], axis=1)
df_encoded.head()
```

```python
numerical_columns = df_encoded.select_dtypes(include=['float64', 'int64']).columns
```

```python
standard_scaler = StandardScaler()
standard_scaled_data = standard_scaler.fit_transform(df_encoded[numerical_columns])
standard_scaled_df = pd.DataFrame(standard_scaled_data, columns=numerical_columns)
standard_scaled_df.head()
```

```python
min_max_scaler = MinMaxScaler()
min_max_scaled_data = min_max_scaler.fit_transform(df_encoded[numerical_columns])
min_max_scaled_df = pd.DataFrame(min_max_scaled_data, columns=numerical_columns)
min_max_scaled_df.head()
```
