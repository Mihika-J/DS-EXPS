```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')
df
```

```python
plt.figure(figsize=(20, 12))
plt.scatter(df['Height'], df['Weight'], c='blue', marker='o')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Scatter Plot')
plt.show()
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.bar(df['Age'],df['Weight'])
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Bar Plot ')
plt.show()
```

```python
plt.hist(df['Age'], bins=5, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Histogram')
plt.show()
```

```python
plt.plot(df['Age'], df['Weight'], marker='o')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Age Over Weight')
plt.grid(True)
plt.show()
```

```python
sns.boxplot(x=df['Weight'], y=df['Age'], data=df)
plt.xlabel('Weight')
plt.ylabel('Age')
plt.title('Box Plot Example')
plt.show()
```
